# scikit-learn中的RBF核(高斯核)
- 对于多项式核函数而言，它的核心思想是将样本数据进行升维，从而使得原本线性不可分的数据线性可分。
- 那么高斯核函数的核心思想是将每一个样本点映射到一个无穷维的特征空间，从而使得原本线性不可分的数据线性可分。

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets

X,y  = datasets.make_moons(noise=0.15,random_state=666)
plt.scatter(X[y==0,0],X[y==0,1],color="r")
plt.scatter(X[y==1,0],X[y==1,1])
plt.show()
```

### Pipeline 与高斯核函数
- 分别测试不同的超参数 gamma = 1,100,10,0.5, 0.1

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC

def RBFKernelSVC(gamma):
    """
        https://www.cnblogs.com/volcao/p/9465214.html
        SVC(kernel = 'ploy')：表示算法使用多项式核函数；
        SVC(kernel = 'rbf')：表示算法使用高斯核函数；
    """
    return Pipeline([
        ("std_scaler",StandardScaler()),
        ("svc",SVC(kernel="rbf",gamma=gamma))
    ])
```

1, 超参数 gamma=1

```python
svc = RBFKernelSVC(gamma=1)
svc.fit(X,y)
```

```python
def plot_decision_boundary(model, axis):
    
    x0, x1 = np.meshgrid(
        np.linspace(axis[0], axis[1], int((axis[1]-axis[0])*100)).reshape(-1, 1),
        np.linspace(axis[2], axis[3], int((axis[3]-axis[2])*100)).reshape(-1, 1),
    )
    X_new = np.c_[x0.ravel(), x1.ravel()]

    y_predict = model.predict(X_new)
    zz = y_predict.reshape(x0.shape)

    from matplotlib.colors import ListedColormap
    custom_cmap = ListedColormap(['#EF9A9A','#FFF59D','#90CAF9'])
    
    plt.contourf(x0, x1, zz, linewidth=5, cmap=custom_cmap)
```

```python
plot_decision_boundary(svc, axis=[-1.5, 2.5, -1.0, 1.5])
plt.scatter(X[y==0,0], X[y==0,1],color="r")
plt.scatter(X[y==1,0], X[y==1,1])
plt.show()
```

2,超参数gamma=100 , 过拟合

```python
svc_gamma100 = RBFKernelSVC(gamma=100)
svc_gamma100.fit(X,y)
```

```python
plot_decision_boundary(svc_gamma100, axis=[-1.5, 2.5, -1.0, 1.5])
plt.scatter(X[y==0,0], X[y==0,1],color="r")
plt.scatter(X[y==1,0], X[y==1,1])
plt.show()
```

3,超参数gamma=10, 

```python
svc_gamma10 = RBFKernelSVC(gamma=10)
svc_gamma10.fit(X,y)
```

```python
plot_decision_boundary(svc_gamma10, axis=[-1.5, 2.5, -1.0, 1.5])
plt.scatter(X[y==0,0], X[y==0,1],color="r")
plt.scatter(X[y==1,0], X[y==1,1])
plt.show()
```

4,超参数gamma=0.5

```python
svc_gamma05 = RBFKernelSVC(gamma=0.5)
svc_gamma05.fit(X, y)
```

```python
plot_decision_boundary(svc_gamma05, axis=[-1.5, 2.5, -1.0, 1.5])
plt.scatter(X[y==0,0], X[y==0,1],color="r")
plt.scatter(X[y==1,0], X[y==1,1])
plt.show()
```

5,超参数gamma=0.1 , 欠拟合

```python
svc_gamma01 = RBFKernelSVC(gamma=0.1)
svc_gamma01.fit(X, y)
```

```python
plot_decision_boundary(svc_gamma01, axis=[-1.5, 2.5, -1.0, 1.5])
plt.scatter(X[y==0,0], X[y==0,1],color="r")
plt.scatter(X[y==1,0], X[y==1,1])
plt.show()
```
