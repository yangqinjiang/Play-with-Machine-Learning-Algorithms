### 11 `matplotlib`基础

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

x = np.linspace(0,10,100) # 生成100个 0至10之间的数字
y = np.sin(x) # sin函数
plt.plot(x,y)
plt.show()
```

```python
siny = y.copy()
```

```python
cosy = np.cos(x)
```

```python
plt.plot(x,siny)
plt.plot(x,cosy)
plt.show()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color="red") # 线条颜色
plt.show()
```

关于color参数：[https://matplotlib.org/2.0.2/api/colors_api.html](https://matplotlib.org/2.0.2/api/colors_api.html)

```python
plt.plot(x,siny)
plt.plot(x,cosy,color="red",linestyle="--") #线条样式
plt.show()
```

关于linestyle参数：[https://matplotlib.org/devdocs/gallery/lines_bars_and_markers/line_styles_reference.html](https://matplotlib.org/devdocs/gallery/lines_bars_and_markers/line_styles_reference.html)

```python
plt.plot(x,siny)
plt.plot(x,cosy,color="red",linestyle="--") #线条样式
# x,y轴的取值范围
plt.xlim(-5,15)
plt.ylim(0,1)
plt.show()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color="red",linestyle="--") #线条样式
# x,y轴的取值范围,另一种方式
plt.axis([-1,11,-2,2]) 
plt.show()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color="red",linestyle="--") #线条样式
# x,y轴的取值范围,另一种方式
plt.axis([-1,11,-2,2]) 
plt.xlabel("x axis")
plt.ylabel("y value")
plt.show()
```

```python
plt.plot(x,siny,label="sin(x)")
plt.plot(x,cosy,color="red",linestyle="--",label="cos(x)") #线条样式
# x,y轴的取值范围,另一种方式
plt.axis([-1,11,-2,2]) 
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend() #图例
plt.show()
```

```python
plt.plot(x,siny,label="sin(x)")
plt.plot(x,cosy,color="red",linestyle="--",label="cos(x)") #线条样式
# x,y轴的取值范围,另一种方式
plt.axis([-1,11,-2,2]) 
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend() #图例
plt.title("Welcome to matplotlib world !")
plt.show()
```

### 散点图

```python
x = np.random.normal(0,1,10000)
y = np.random.normal(0,1,10000)

plt.scatter(x,y,alpha=0.1) #alpha 透明度
plt.show()
```
