## kNN的决策边界
对于kNN算法来说, k越小,决策边界越不规则, 越过拟合

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data[:,:2] # 前两个特征
y = iris.target
```

画图表示前两个特征

```python
plt.scatter(X[y==0,0],X[y==0,1])
plt.scatter(X[y==1,0],X[y==1,1])
plt.scatter(X[y==2,0],X[y==2,1])
plt.show()
```

```python
def plot_decision_boundary(model, axis):
    
    x0, x1 = np.meshgrid(
        np.linspace(axis[0], axis[1], int((axis[1]-axis[0])*100)).reshape(-1, 1),
        np.linspace(axis[2], axis[3], int((axis[3]-axis[2])*100)).reshape(-1, 1),
    )
    X_new = np.c_[x0.ravel(), x1.ravel()]

    y_predict = model.predict(X_new)
    zz = y_predict.reshape(x0.shape)

    from matplotlib.colors import ListedColormap
    custom_cmap = ListedColormap(['#EF9A9A','#FFF59D','#90CAF9'])
    
    plt.contourf(x0, x1, zz, linewidth=5, cmap=custom_cmap)
```

### 当K=1时, 边界不规则, 过拟合, 但score很高

```python
from sklearn.neighbors import KNeighborsClassifier
knn_clf_1 = KNeighborsClassifier(n_neighbors=1)
knn_clf_1.fit(X,y)
print(knn_clf_1.score(X,y)) 
plot_decision_boundary(knn_clf_1, axis=[4, 8, 1.5, 4.5])
plt.scatter(X[y==0,0], X[y==0,1])
plt.scatter(X[y==1,0], X[y==1,1])
plt.scatter(X[y==2,0], X[y==2,1])
plt.show()
```

### 当K=5时, 边界不规则, 过拟合, 但比K=1的好, score比k=1的低

```python
knn_clf_5 = KNeighborsClassifier(n_neighbors=5)
knn_clf_5.fit(X, y)
print(knn_clf_5.score(X,y))
plot_decision_boundary(knn_clf_5, axis=[4, 8, 1.5, 4.5])
plt.scatter(X[y==0,0], X[y==0,1])
plt.scatter(X[y==1,0], X[y==1,1])
plt.scatter(X[y==2,0], X[y==2,1])
plt.show()
```

### 当K=25时, 边界不规则, 过拟合, 但比K=5的好, score比k=5的低

```python
knn_clf_25 = KNeighborsClassifier(n_neighbors=25)
knn_clf_25.fit(X, y)
print(knn_clf_25.score(X,y))
plot_decision_boundary(knn_clf_25, axis=[4, 8, 1.5, 4.5])
plt.scatter(X[y==0,0], X[y==0,1])
plt.scatter(X[y==1,0], X[y==1,1])
plt.scatter(X[y==2,0], X[y==2,1])
plt.show()
```

### 当K=100时, 边界比较规则, score比k=25的低

```python
knn_clf_100 = KNeighborsClassifier(n_neighbors=100)
knn_clf_100.fit(X, y)
print(knn_clf_100.score(X,y))

plot_decision_boundary(knn_clf_100, axis=[4, 8, 1.5, 4.5])
plt.scatter(X[y==0,0], X[y==0,1])
plt.scatter(X[y==1,0], X[y==1,1])
plt.scatter(X[y==2,0], X[y==2,1])
plt.show()
```
